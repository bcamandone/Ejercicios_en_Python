Vamos a jugar al bingo!!

Escribe un programa que solicite al usuario un número de jugadores y sus respectivos nombres. 

Generar cartones para cada uno de los jugadores y mostrarlos por pantalla. Cada cartón tiene 15 números-del 1 al 90- divididos en tres líneas de 5. Tener en cuenta que los números no pueden aparecer repetidos. 

Una vez establecidos los números, se irán cantando las bolas y se irán tachando los números de los cartones. Anunciar el ganador de la primera línea y el ganador final.

```python
def element_removal(lista,value):
  lista_nueva = []
  for elemento in lista:
    if elemento != value:
      lista_nueva.append(elemento)
  return lista_nueva
```

```python
def generar_carton():
  #El cartón se guardará en una lista de listas
  numeros = [i for i in range(1,91)]
  carton = []
  #el primer bucle es para generar cada linea
  for i in range(3):
    linea = []
    #El segundo bucle es para generar los numeros dentro de una linea
    for j in range(5):
      #sacamos un numero de los que hay en la lista de numeros
      numero = elemento_aleatorio(numeros)
      #una vez sacado el numero, lo quitamos de la lista para que no vuelva a salir
      numeros = element_removal(numeros, numero)
      #metemos el numero en la linea
      linea.append(numero)
    #Una vez generada toda la linea, la metemos en el carton
    carton.append(linea)
  return carton
```

```python
from random import randint
def elemento_aleatorio(lista):
  #sacamos un indice aleatorio
  indice_elemento = randint(1,len(lista))-1
  return lista[indice_elemento]
```

```python
def bingo():
  #Pedimos numero de jugadores 
  num_jugadores = int(input('Por favor introduce el numero de jugadores:'))
  lista_jugadores = []
  #Pedimos tantas veces nombre de jugadores como jugadores haya
  for i in range(num_jugadores):
    #Ponemos i+1 para que no imprima el numero 0
    nombre_jugador = input(f'Ingresa el nombre del jugador {i+1}:')
    lista_jugadores.append(nombre_jugador)
  print('Se generarán ahora los cartones:')
  #El diccionario es perfecto para guardar cada cartón, así podemos vincularlo con los nombres de los jugadores
  dict_cartones = {}
  for nombre in lista_jugadores:
    carton = generar_carton()
    #guardamos el carton de cada jugador
    dict_cartones[nombre] = carton
    #Con \n saltamos de linea
    print(f"Este es el carton de {nombre}: \n")
    print(carton)
  print('Empezamos a jugar!')
  numeros = [i for i in range (1,91)]
  hay_linea = False
  hay_ganador = False
  #mientras no haya ganador
  while hay_ganador == False:
    #se saca un numero
    numero = elemento_aleatorio(numeros)
    #imprimimos en pantalla para no perder el control de lo que sale
    print(numero)
    # sacamos el numero del listado para que no salga de nuevo
    numeros = element_removal(numeros, numero)
    for nombre in lista_jugadores:
      carton_jugador = dict_cartones[nombre]
      #para cada linea de cada carton
      for i in range(3):
        #quitamos el numero de esa linea
        carton_jugador[i] = element_removal(carton_jugador[i],numero)
        #una vez quitado el numero de la linea, y si no se ha cantado todavía
        if hay_linea == False:
          #vemos si la línea se ha quedado sin numeros
          if len(carton_jugador[i]) == 0:
            #cantamos la linea
            print(f'El ganador de la linea es {nombre}!!')
            #ponemos que hay linea para que no la cante más
            hay_linea = True
      #guardamos el carton modificado del jugador
      dict_cartones[nombre] = carton_jugador
      #comprobamos si las tres lineas están vacias
      if len(carton_jugador[0])+len(carton_jugador[1])+len(carton_jugador[2]) == 0:
        #cambiamos hay_ganador para que pare de ejecutarse el while
        hay_ganador = True
        print(f'{nombre}, has ganado!')
```
